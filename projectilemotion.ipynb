```python
import math
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
vo = [60]
theta = [10,20,30,40,45,50,60,70,80,85]
```

first function calculates x component of the initial velocity (meters/second)

```python
def componentx(vo, theta):
    theta = theta * (math.pi / 180)
    xcomp = vo * math.cos(theta)
    return xcomp
```

second function calculates y component of the initial velocity

```python
def componenty(vo, theta):
    theta = theta * (math.pi / 180)
    ycomp = vo * math.sin(theta)
    return ycomp
```

third function takes the y component and calculates the flight time (seconds)

```python
def flighttime(ycomp):
    time = ((0 - ycomp) / (-9.8)) * 2
    return time
```

fourth function takes the time (seconds) and calculates y displacement (meters)

```python
def ydisp(ycomponentvo, time):
    ydisplacement = (ycomponentvo * time) + (0.5 * -9.8 * (time**2))
    return ydisplacement
```

fifth function takes time and calculates x displacement 

```python
def xdisp(xcomponentvo, time):
    xdisplacement = (xcomponentvo*time)
    return xdisplacement                                                     
```

first method was to try and create a dataframe then plot

```python
biglist = []
for angle in theta:
    dictionary = {}
    for iteration in np.arange(0,25,0.5):   
        x = xdisp(componentx(vo[0], angle), iteration)
        y = ydisp(componenty(vo[0], angle), iteration)
        dictionary[str(iteration)] = tuple((x,y))
    biglist.append(dictionary)
       
   

dataframe = pd.DataFrame(biglist, theta)
dataframe
```

then made a function to calculate the displacement at different launch angles and added all these values to a list

```python
def xycalc(launchangle):
    xs = []
    ys = []
    for iteration in np.arange(0,30,0.5):   
        x = xdisp(componentx(vo[0], launchangle), iteration)
        y = ydisp(componenty(vo[0], launchangle), iteration)
        if y >= 0:
            xs.append(x)
            ys.append(y)
        else: #to find the final coordinate where (x,0)
            xfinal = xdisp(componentx(vo[0], launchangle), flighttime(componenty(vo[0], launchangle)))
            yfinal = 0
            xs.append(xfinal)
            ys.append(yfinal)
            break
    return xs, ys
```

plot (x,y) for each launch angle

```python
for angle in theta:
   xs, ys = xycalc(angle)
   plt.plot(xs, ys, label = str(angle)+" degrees")
```

```python
plt.rcParams["figure.figsize"] = [11,8]
plt.xlabel("Horizontal Displacement (m)")
plt.ylabel("Vertical Displacement (m)")
plt.title("Projectile Motion at Various Angles")
plt.legend()
plt.show()

```
